```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
dataset = "C:\\Users\\X474241\\OneDrive - Old Mutual\\Documents\\GitHub\\describing_distributions\\Data\\data (1).csv"
df = pd.read_csv(dataset, index_col = 0)
df.head()
```

```python
df.info()
```

```python
# plotting a histogram of VitD_levels 
sns.histplot(data=df,x='VitD_levels',kde =True)
```

```python
#Add a line for the mean (red)
mean = df['VitD_levels'].mean()
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(df['VitD_levels'], ax = ax, stat='probability')
ax.axvline(mean, color = 'red', label = 'mean');
ax.set_title('VitD_levels')
ax.legend();
```

```python
#Add a line for the median (green)
col = 'VitD_levels'
feature = df[col]
mean = feature.mean()
median = feature.median()
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(mean, color = 'red',label=f'Mean = {mean:,.2f}')
ax.axvline(median, ls='--', color = 'green', label = f'Median = {median:,.2f}')
ax.set_title(f'{col}')
ax.legend();
```

```python
#Add a line for for +1 std from the mean (black)
col = 'VitD_levels'
feature = df[col]
mean = feature.mean()
std = feature.std()
plus_one_std = mean + std
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.set_title(f'{col}')
ax.legend()
```

```python
#Add a line for the - 1 std from the mean (black)
col = 'VitD_levels'
feature = df[col]
mean = feature.mean()
std = feature.std()
minus_one_std = mean - std
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.set_title(f'{col}')
ax.legend();
```

```python
#Highlight the range between +1 and =1 std (yellow)
col = 'VitD_levels'
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.set_title(f'{col}')
ax.legend();
```

# What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)

```python
# VitD_levels displays Mesokurtic
kurt_VitD_levels = stats.kurtosis(df['VitD_levels'], fisher = False)
kurt_VitD_levels
```

```python
# Doc_visits displays Mesokurtic
kurt_Doc_visits = stats.kurtosis(df['Doc_visits'], fisher = False)
kurt_Doc_visits
```

```python
# TotalCharge displays Leptokurtic
kurt_TotalCharge = stats.kurtosis(df['TotalCharge'], fisher = False)
kurt_TotalCharge
```
